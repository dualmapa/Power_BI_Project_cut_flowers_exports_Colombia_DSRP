```python
import zipfile
import pandas as pd
import os

# Ruta del archivo ZIP
zip_path = '/content/datos_exportaciones.zip'

# Abrir archivo ZIP
with zipfile.ZipFile(zip_path, 'r') as z:
    # Listar archivos CSV en el ZIP
    csv_files = [name for name in z.namelist() if name.endswith('.csv')]

    # Inicializar un set para almacenar estructuras de columnas
    columnas_unicas = set()

    # Revisar cada archivo
    for file_name in csv_files:
        with z.open(file_name) as f:
            try:
                df = pd.read_csv(f, nrows=5, encoding='latin-1', sep=None, engine='python')
                columnas_unicas.add(tuple(df.columns))
            except Exception as e:
                print(f"Error leyendo {file_name}: {e}")

# Resultados
if len(columnas_unicas) == 1:
    print("Todos los archivos tienen la misma estructura.")
else:
    print("Hay inconsistencias en las columnas:")
    for cols in columnas_unicas:
        print(cols)
```

```python
import zipfile
import pandas as pd
import os

# Ruta del archivo ZIP
zip_path = '/content/datos_exportaciones.zip'

# Abrir archivo ZIP
with zipfile.ZipFile(zip_path, 'r') as z:
    # Listar archivos CSV en el ZIP
    csv_files = [name for name in z.namelist() if name.endswith('.csv')]

    # Diccionario para almacenar columnas por archivo
    columnas_por_archivo = {}

    # Revisar cada archivo
    for file_name in csv_files:
        with z.open(file_name) as f:
            try:
                # Leer primeras filas para identificar columnas
                df = pd.read_csv(f, nrows=5, encoding='latin-1', sep=None, engine='python')

                # Limpiar nombres de columnas (eliminar caracteres extraños y espacios)
                df.columns = df.columns.str.replace('ï»¿', '', regex=False)  # Eliminar BOM
                df.columns = df.columns.str.strip()  # Eliminar espacios al inicio y final

                # Eliminar las columnas 'NIT' y 'RAZ_SIAL' si existen
                df = df.loc[:, ~df.columns.isin(['NIT', 'RAZ_SIAL'])]

                # Almacenar la estructura de las columnas después de la limpieza
                columnas_por_archivo[file_name] = tuple(df.columns)

            except Exception as e:
                print(f"Error leyendo {file_name}: {e}")

# Analizar las estructuras de columnas
estructura_unica = {}
for archivo, columnas in columnas_por_archivo.items():
    if columnas not in estructura_unica:
        estructura_unica[columnas] = []
    estructura_unica[columnas].append(archivo)

# Mostrar resultados
for columnas, archivos in estructura_unica.items():
    print(f"Estructura de columnas: {columnas}")
    print(f"Archivos que la tienen ({len(archivos)} archivos):")
    for archivo in archivos:
        print(f" - {archivo}")
    print("\n" + "-"*80 + "\n")
```

```python
import zipfile
import pandas as pd
import os

# Ruta del archivo ZIP
zip_path = '/content/datos_exportaciones.zip'

# Lista para almacenar los DataFrames de cada archivo
all_data = []

with zipfile.ZipFile(zip_path, 'r') as z:
    # Listar archivos CSV en el ZIP
    csv_files = [name for name in z.namelist() if name.endswith('.csv')]

    for file_name in csv_files:
        with z.open(file_name) as f:
            try:
                # Leer el CSV con las columnas limpiadas
                df = pd.read_csv(f, encoding='latin-1', sep=None, engine='python')
                df.columns = df.columns.str.replace('ï»¿', '', regex=False)  # Eliminar BOM
                df.columns = df.columns.str.strip()  # Eliminar espacios al inicio y final
                df = df.loc[:, ~df.columns.isin(['NIT', 'RAZ_SIAL'])]  # Eliminar columnas no deseadas

                all_data.append(df)  # Añadir el DataFrame a la lista

            except Exception as e:
                print(f"Error leyendo {file_name}: {e}")

# Combina todos los DataFrames en uno solo
consolidated_data = pd.concat(all_data, ignore_index=True)

# Asegurarse de que la columna 'POSAR' sea de tipo string antes de aplicar el filtro
consolidated_data['POSAR'] = consolidated_data['POSAR'].astype(str)

# Filtrar solo los valores que empiecen con los códigos indicados
filtered_data = consolidated_data[consolidated_data['POSAR'].str.startswith(('06031', '6031', '06042', '6042'))]

# Mostrar las primeras filas del DataFrame filtrado
print(filtered_data.head())
```
